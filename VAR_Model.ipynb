```python
import pandas as pd
import numpy as np
```

```python
from statsmodels.tsa.stattools import grangercausalitytests
from statsmodels.tsa.stattools import adfuller
import statsmodels.api as sm
from statsmodels.tsa.api import VAR
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['figure.figsize'] = (10,8)
mpl.rcParams['axes.grid'] =  False
```

```python
df = pd.read_csv('appliance_energy_usage.csv',index_col=0,parse_dates=True)
```

```python
df.head()
```

```python
df = df.resample('1H').mean() #for sake of visualisations
```

```python
color_list = ["blue","orange","green","red","purple","brown","pink","gray","olive","cyan"]
```

```python
def viz(data):
    features = list(df.select_dtypes(include=[np.number]).columns.values) #selecting columns with numerical values
    feature_size = len(features)
    fig,axes = plt.subplots(
        nrows=int(np.ceil(feature_size/2)),ncols=2,figsize=(14,feature_size*2),dpi=80,facecolor="w",edgecolor="k"
    )
    for i in range(feature_size):
        key = features[i]
        c = color_list[i%(len(color_list))]
        t_data = data[key]
        t_data.head()
        ax = t_data.plot(
            ax = axes[i//2,i%2],
            color = c,
            title = "{}".format(key),
            rot = 25,
    )
    ax.legend(key)
    plt.tight_layout()
    
viz(df)
```

```python
for i in range(len(df.columns)):
    result = adfuller(df[df.columns[i]],autolag = "BIC")
    print('{} - Series test-statistics is {} and critical value at 5% significance level is {}'.format(df.columns[i],result[0],result[4]['5%']))
    print ('{} - Series p-value is {} and used lag is {}'.format(df.columns[i],result[1],result[2]))
    
    if result[1] > 0.05 :
        print('{} -Series is not Stationary'.format(df.columns[i]))
    else:
        print('{} -Series is Stationary'.format(df.columns[i]))
```

# GrangerCausalityTest

```python
max_lags = 8 
y = 'Appliances'
```

```python
for i  in range(len(df.columns)-1):
    results = grangercausalitytests(df[[y,df.columns[i+1]]],max_lags,verbose=False)
    p_values= [round(results[i+1][0]['ssr_ftest'][1],4) for i in range(max_lags)]
    print('Column - {} : P_values - {}'.format(df.columns[i+1],p_values))
    
```
